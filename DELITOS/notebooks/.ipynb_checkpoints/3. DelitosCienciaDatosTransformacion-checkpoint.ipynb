## Cantidad total de delitos cometidos durante todo el año.

```python
delitos_df['TotalDelitos'] = delitos_df.iloc[:, 7:19].sum(axis=1)
delitos_df
```

##  Agrupamiento de los delitos por categorías más amplias (por ejemplo, "delitos contra la propiedad" y "delitos contra las personas").

```python
def categorizar_delito(tipo_delito):
    palabras_contra_propiedad = ['robo', 'hurto', 'fraude', 'extorsión', 'daño', 'allanamiento']
    palabras_contra_personas = ['homicidio', 'lesiones', 'violación', 'abuso', 'secuestro', 'amenaza']
    if any(palabra in tipo_delito.lower() for palabra in palabras_contra_propiedad):
        return 'Delitos contra la propiedad'
    elif any(palabra in tipo_delito.lower() for palabra in palabras_contra_personas):
        return 'Delitos contra las personas'
    else:
        return 'Otros delitos'
    
delitos_df['Categoria de delito'] = delitos_df['Tipo de delito'].apply(categorizar_delito)
delitos_df

```

## Columna con el mes con mayor delito

```python
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
mes_max = delitos_df[meses].idxmax(axis=1)
delitos_df['Mes con mayor delitos'] = mes_max
delitos_df
```
