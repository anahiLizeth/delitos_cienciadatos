## ANALISIS MULTIDIMENSIONAL DE LAS VARIABLES

```python
#suma totak de los delitos cometidos por año

# Definir el rango de columnas para sumar
meses_a_sumar = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
# Agregar una nueva columna que sea la suma de las columnas seleccionadas
delitos_df['TotalPorAño'] = delitos_df[meses_a_sumar].sum(axis=1)
delitos_df.head(5)
```

```python
suma_delitos_df_entidad = delitos_df.groupby('Entidad').sum()['TotalPorAño']
suma_delitos_df_entidad
```
