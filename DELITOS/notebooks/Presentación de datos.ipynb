```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import nltk

from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Lee los datos del archivo CSV
delitos_df = pd.read_csv('clean_final_delitos.csv')

```

## GRÁFICA QUE MUESTRA EL NUMERO DE DELITOS POR AÑO Y POR TIPO

```python

# Agrupar los delitos por año y tipo, y sumar los totales
delitos_agrupados = delitos_df.groupby(['Año', 'Tipo de delito'])['TotalDelitos'].sum().reset_index()

# Calcular los totales por año
totales_por_año = delitos_agrupados.groupby('Año')['TotalDelitos'].sum()

# Ordenar los datos por el total de delitos
delitos_agrupados = delitos_agrupados.sort_values(by=['Año', 'TotalDelitos'], ascending=[True, False])

# Seleccionar los tipos de delito con los mayores totales
tipos_de_delito = delitos_agrupados.groupby('Año').head(5)['Tipo de delito'].unique()

# Filtrar los datos para incluir solamente los tipos de delito seleccionados
delitos_filtrados = delitos_agrupados[delitos_agrupados['Tipo de delito'].isin(tipos_de_delito)]

# Crear una gráfica de barras que muestre el total de delitos agrupados por año y tipo de delito
for tipo_de_delito in tipos_de_delito:
    datos_por_tipo = delitos_filtrados[delitos_filtrados['Tipo de delito'] == tipo_de_delito]
    plt.bar(datos_por_tipo['Año'], datos_por_tipo['TotalDelitos'], label=tipo_de_delito)

plt.legend()
plt.xlabel('Año')
plt.ylabel('Total de delitos')
plt.show()



```

## NUBE DE PALABRAS QUE SE REPITEM MAS EN LOS TIPOS DE DELITOS

```python
text_delitos =''
for row in delitos_df['Tipo de delito']:
    text_delitos = text_delitos+' '+row

stopwords_list = stopwords.words('spanish')+['2023','cont','millar','dpp']
wc = WordCloud(background_color = 'white')
wc.generate(text_delitos)

plt.axis("off")
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

## NUMERO DE DELITOS POR AÑO

```python

df_agrupado = delitos_df.groupby(['Año', 'Mes con mayor delitos'])['TotalDelitos'].sum().reset_index()

df_mes_mayor = df_agrupado.loc[df_agrupado.groupby('Año')['TotalDelitos'].idxmax()]

plt.bar(df_mes_mayor['Año'], df_mes_mayor['TotalDelitos'])
plt.title('Número de delitos por año y mes con mayor delitos')
plt.xlabel('Año')
plt.ylabel('Número de delitos')
plt.show()
```

## MES CON MAYOR NUMERO DE DELITOS POR AÑO

```python

# Agrupar por año y mes, y sumar los delitos
df_grouped = delitos_df.groupby(["Año", "Mes con mayor delitos"]).agg({"TotalDelitos": "sum"}).reset_index()

# Encontrar el mes con el número máximo de delitos para cada año
df_max_month = df_grouped.loc[df_grouped.groupby("Año")["TotalDelitos"].idxmax()]

# Agregar una columna con la combinación de año y mes
df_max_month["Year-Month"] = df_max_month["Año"].astype(str) + "-" + df_max_month["Mes con mayor delitos"].astype(str)

# Crear la gráfica de barras
plt.bar(df_max_month["Year-Month"], df_max_month["TotalDelitos"])
# Personalizar las etiquetas del eje x
plt.xticks(rotation=45, ha='right', fontsize=8)
# Establecer los títulos y etiquetas de los ejes
plt.title("Mes con mayor número de delitos por año")
plt.xlabel("Año-Mes")
plt.ylabel("Número de delitos")

# Mostrar la gráfica
plt.show()
```
